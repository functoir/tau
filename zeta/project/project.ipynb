```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
pd.set_option('display.max_rows', 600)
from pathlib import Path  
import glob
from os.path import exists as file_exists
```

```python
# stop words

stop_words = []
with open("../data/metadata/stopwords.txt") as f:
  for line in f:
    # print(line)
    stop_words.append(line.strip()) # .append(line.strip())

print("Hello")
print(f"{stop_words = }")
```

```python
# years to focus on
years = list(range(1990, 2023))

# max count per year
max_count = 10

years, max_count
```

1.

```python
# load $count files for each year.
directory_path = "../data/categorized"
all_words = []

for year in years:
  files = [f for f in glob.glob(f"{directory_path}/{year}/*.txt") if file_exists(f)]

  words = ""
  for file in files:
    with open(file) as f:
      for line in f:
        words += line

  all_words.append(words)
  # titles = [Path(text).stem for text in files]

tfidf_vectorizer = TfidfVectorizer(input='content', stop_words=stop_words, strip_accents="ascii")
tfidf_vector = tfidf_vectorizer.fit_transform(all_words)
del all_words

tfidf_df = pd.DataFrame(tfidf_vector.toarray(), index=[str(year) for year in years], columns=tfidf_vectorizer.get_feature_names_out())
# data.append(tfidf_df)
data = tfidf_df
data.loc["document"] = data.sum(axis=0)
```

```python
# peek at our data
data
```

```python
# Re-organize so words are in rows rather than columns
cleaned_data = data.sort_index().stack().reset_index()
cleaned_data = cleaned_data.rename(columns={0:'tfidf', 'level_0': 'year','level_1': 'term', 'level_2': 'term'})
# stacked_tfidf_df.sample(n=20)
cleaned_data.sample(n=20)
```

```python
# top 10 words for all stories
#! MemoryError: Unable to allocate 97.5 GiB for an array with shape (23351, 560214) and data type float64
top_tfidf = cleaned_data.groupby(['year', 'term'])['tfidf'].sum().sort_values(ascending=False).head(max_count)
top_tfidf
```
